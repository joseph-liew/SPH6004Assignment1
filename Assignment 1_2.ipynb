**mann whitney u test for contiuous variables**

```python
import pandas as pd
import numpy as np
import scipy as sp

data_missing_less_40p = pd.read_csv('Assignment_1_data_missing_less_40p.csv')

pd.set_option("display.max_columns", None)


noncontinuous_var = ['gender', 'sofa_coagulation', 'sofa_cardiovascular','sofa_cns', 'sofa_renal', 'charlson_comorbidity_index']

data_continuous = data_missing_less_40p.loc[:, ~data_missing_less_40p.columns.isin(noncontinuous_var)]



select = data_continuous[['age', 'outcome']]
##age_outcome = select[select['outcome'] == True]
age_true = select[select['outcome'] == True]['age']
age_true = pd.DataFrame(data=age_true).T
age_true = age_true.loc['age'].values.flatten().tolist()

select = data_continuous[['age', 'outcome']]
#age_outcome = select[select['outcome'] == False]
age_false = select[select['outcome'] == False]['age']
age_false = pd.DataFrame(data=age_false).T
age_false = age_false.loc['age'].values.flatten().tolist()

age_mann = sp.stats.mannwhitneyu(age_true, age_false, use_continuity=True, nan_policy='omit')

#print('age mann whitney outcome:', age_mann)
#print('\n')

col_names = data_continuous.loc[:, ~data_continuous.columns.isin(['outcome'])]
col_names = list(col_names)


def describeMann(data_table, variable, treatment, treat_label1, treat_label2):
    select = data_table[[variable, treatment]]
    var_true = select[select[treatment] == treat_label1][variable]
    var_true = pd.DataFrame(data=var_true).T
    var_true = var_true.loc[variable].values.flatten().tolist()

    select2 = data_table[[variable, treatment]]
    var_false = select2[select2[treatment] == treat_label2][variable]
    var_false = pd.DataFrame(data=var_false).T
    var_false = var_false.loc[variable].values.flatten().tolist()

    var_mann = sp.stats.mannwhitneyu(var_true, var_false, use_continuity=True, nan_policy='omit')
    
    return var_mann

for i in col_names:
    print('The ' + i + ' ' + str(describeMann(data_continuous, i, 'outcome', True, False)))
    print('\n')
```

**non_continuous chi sq outcome**

```python
import pandas as pd
import numpy as np
import scipy as sp

data_missing_less_40p = pd.read_csv('Assignment_1_data_missing_less_40p.csv')

pd.set_option("display.max_columns", None)


noncontinuous_var = ['gender','sofa_coagulation', 'sofa_cardiovascular','sofa_cns', 'sofa_renal', 'charlson_comorbidity_index']

def chisq(data_table, variable, treatment):
    contig = pd.crosstab(data_table[variable], data_table[treatment], rownames=None, colnames=None)
    
    chi2_contig = sp.stats.chi2_contingency(contig, lambda_=None, correction=False)
    return chi2_contig # chi2_contingency() returns chi(which is a test statistic), p(which is the p-value of the test), dof(which is the degree of freedom) and expected(Frequencies to expect)

for i in noncontinuous_var:
    print('The ' + i + ' chi2 contingency is p-value: ' + str(chisq(data_missing_less_40p, i, 'outcome')[1]) + ' details: chi / p / dof / expected '+ str(chisq(data_missing_less_40p, i, 'outcome')))
    print('\n')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier, SGDRegressor, RidgeCV, LassoCV
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder, StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.impute import KNNImputer, SimpleImputer
from sklearn_genetic import GAFeatureSelectionCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.metrics import classification_report, roc_auc_score
from collections import Counter
from imblearn.over_sampling import SMOTE, SMOTEN

print('SGDClassifier')
print()
smoten_noncontin = pd.read_csv('smoten_noncontin.csv', index_col=False)

X = smoten_noncontin.drop('outcome',axis= 1)
y = smoten_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

estimate = SGDClassifier(loss='hinge', penalty='l2', alpha=0.0001, l1_ratio=0.15, fit_intercept=True, max_iter=1000, tol=0.001, shuffle=True, verbose=0, epsilon=0.1, n_jobs=None, random_state=None, learning_rate='optimal', eta0=0.0, power_t=0.5, early_stopping=False, validation_fraction=0.1, n_iter_no_change=5, class_weight=None, warm_start=False, average=False)

cv = 5 
top_k = 1
dispatch = '8*n_jobs'
return_train = True
gen = 40
geneSelectFeature = GAFeatureSelectionCV(estimate, cv=cv, scoring=None, population_size=50, generations=gen, crossover_probability=0.2, mutation_probability=0.8, tournament_size=3, elitism=True, max_features=None, verbose=True, keep_top_k=top_k, criteria='max', algorithm='eaMuPlusLambda', refit=True, n_jobs=1, pre_dispatch=dispatch, error_score=np.nan, return_train_score=return_train, log_config=None)

print('Genetic Algorithm Feature Selection StandardScaler Non-Contin SMOTEN')
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Algorithm Feature Selection RobustScaler Non-Contin SMOTEN')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

knn_imputed_noncontin = pd.read_csv('knn_imputed_noncontin.csv', index_col=False)

X = knn_imputed_noncontin.drop('outcome',axis= 1)
y = knn_imputed_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

print('Genetic Feature Selection StandardScaler Non-Contin Encoded' )
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Feature Selection RobustScaler Non-Contin Encoded')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')


print('SVC(kernel = rbf, C=1.0)')
print()

X = smoten_noncontin.drop('outcome',axis= 1)
y = smoten_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

estimate = SVC(kernel='rbf', C=1.0)

geneSelectFeature = GAFeatureSelectionCV(estimate, cv=cv, scoring=None, population_size=50, generations=gen, crossover_probability=0.2, mutation_probability=0.8, tournament_size=3, elitism=True, max_features=None, verbose=True, keep_top_k=top_k, criteria='max', algorithm='eaMuPlusLambda', refit=True, n_jobs=1, pre_dispatch=dispatch, error_score=np.nan, return_train_score=return_train, log_config=None)

print('Genetic Algorithm Feature Selection StandardScaler Non-Contin SMOTEN')
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Algorithm Feature Selection RobustScaler Non-Contin SMOTEN')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

X = knn_imputed_noncontin.drop('outcome',axis= 1)
y = knn_imputed_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

print('Genetic Feature Selection StandardScaler Non-Contin Encoded' )
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Feature Selection RobustScaler Non-Contin Encoded')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')



print('SVC(kernel = poly, C=1.0)')
print()
smoten_noncontin = pd.read_csv('smoten_noncontin.csv', index_col=False)

X = smoten_noncontin.drop('outcome',axis= 1)
y = smoten_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

estimate = SVC(kernel='poly', C=1.0)

geneSelectFeature = GAFeatureSelectionCV(estimate, cv=cv, scoring=None, population_size=50, generations=gen, crossover_probability=0.2, mutation_probability=0.8, tournament_size=3, elitism=True, max_features=None, verbose=True, keep_top_k=top_k, criteria='max', algorithm='eaMuPlusLambda', refit=True, n_jobs=1, pre_dispatch=dispatch, error_score=np.nan, return_train_score=return_train, log_config=None)

print('Genetic Algorithm Feature Selection StandardScaler Non-Contin SMOTEN')
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Algorithm Feature Selection RobustScaler Non-Contin SMOTEN')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

X = noncontinuous_encoded.drop('outcome',axis= 1)
y = noncontinuous_encoded['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

print('Genetic Feature Selection StandardScaler Non-Contin Encoded' )
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Feature Selection RobustScaler Non-Contin Encoded')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')


print('SVC(kernel = rbf, C=1.0)')
print()

X = smoten_noncontin.drop('outcome',axis= 1)
y = smoten_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

estimate = SVC(kernel='rbf', C=1.0)

geneSelectFeature = GAFeatureSelectionCV(estimate, cv=cv, scoring=None, population_size=50, generations=gen, crossover_probability=0.2, mutation_probability=0.8, tournament_size=3, elitism=True, max_features=None, verbose=True, keep_top_k=top_k, criteria='max', algorithm='eaMuPlusLambda', refit=True, n_jobs=1, pre_dispatch=dispatch, error_score=np.nan, return_train_score=return_train, log_config=None)

print('Genetic Algorithm Feature Selection StandardScaler Non-Contin SMOTEN')
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneticFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Algorithm Feature Selection RobustScaler Non-Contin SMOTEN')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneticFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

X = knn_imputed_noncontin.drop('outcome',axis= 1)
y = knn_imputed_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

print('Genetic Feature Selection StandardScaler Non-Contin Encoded' )
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Feature Selection RobustScaler Non-Contin Encoded')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('SVC(kernel = sigmoid, C=1.0)')
print()

X = smoten_noncontin.drop('outcome',axis= 1)
y = smoten_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

estimate = SVC(kernel='sigmoid', C=1.0)

geneSelectFeature = GAFeatureSelectionCV(estimate, cv=cv, scoring=None, population_size=50, generations=gen, crossover_probability=0.2, mutation_probability=0.8, tournament_size=3, elitism=True, max_features=None, verbose=True, keep_top_k=top_k, criteria='max', algorithm='eaMuPlusLambda', refit=True, n_jobs=1, pre_dispatch=dispatch, error_score=np.nan, return_train_score=return_train, log_config=None)

print('Genetic Algorithm Feature Selection StandardScaler Non-Contin SMOTEN')
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Algorithm Feature Selection RobustScaler Non-Contin SMOTEN')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

X = knn_imputed_noncontin.drop('outcome',axis= 1)
y = knn_imputed_noncontin['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = 0) #train split

print('Genetic Feature Selection StandardScaler Non-Contin Encoded' )
clf = make_pipeline(StandardScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf[1].predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')

print('Genetic Feature Selection RobustScaler Non-Contin Encoded')
clf = make_pipeline(RobustScaler(), geneSelectFeature)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('best estimator ' + str(clf[1].best_estimator_))
outcome_labels = ['Intubation False', 'Intubation True']

print('Genetic Feature Selection:', geneSelectFeature.support_, '\n')
print('score \n', clf.score(X_train, y_train))
print('classification report \n', classification_report(y_test, y_pred, target_names=outcome_labels))
micro_roc_auc_ovr = roc_auc_score(y_test, y_pred, multi_class="ovr", average="micro")
print(f"Micro-averaged One-vs-Rest ROC AUC score:\n{micro_roc_auc_ovr:.2f}")
print('\n')
```
